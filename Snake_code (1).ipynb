```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
#code 1
pos = np.zeros([2], int)
length = 400
record = np.zeros([length+1,2],int)
print(record[0,:])
size = 0
def game(iterations, pos, size):
    j = 0
    count = 0
    i = 0
    while i < iterations:
        if j == 0:
            x = random.randint(1, 4)
        else:
            x = x % 4 + 1 
        if (x == 1):
            pos[0] = record[i,0] + 1
        elif (x == 2):
            pos[0] = record[i,0] - 1
        elif (x == 3):
            pos[1] = record[i,1] + 1
        else:
            pos[1] = record[i,1] - 1
        for k in range(i):
            if (pos[0] == record[k,0] and pos[1] == record[k,1] and i !=0):
                j += 1
                if j == 4:
                    size = i + 1
                    #print('trapped with a length of', size) ## works for this code but not with code 3
                    return size
                pos[0] = record[i,0]
                pos[1] = record[i,1]  
                break
            if (k == i-1):
                j = 0
        if (j == 0):
            record[i+1,0] = pos[0]
            record[i+1,1] = pos[1]
            j = 0
            #print(record[i+1,:]) ## same for this line of code
            i += 1

size = game(length, pos, size)
```

```python
#code 2
import matplotlib.pyplot as plt
if type(size) == int:
    x = record[0:size,0]
    y = record[0:size,1]
else:
    x = record[:,0]
    y = record[:,1]
plt.plot(x, y, 'o-')
plt.plot(0, 0, 'o-', color = 'lime')
if type(size) == int:
    plt.plot(record[size-1,0], record[size-1,1], 'o-', color = 'red')
#plt.xticks(np.arange(-maximum -1, maximum + 2, gap))
#plt.yticks(np.arange(-maximum -1 , maximum + 2, gap))
plt.xlim(min(record[:,0])-1,max(record[:,0])+1)
plt.ylim(min(record[:,1])-1,max(record[:,1])+1)
plt.show()

if type(size) == int:
    print('the length of the chain is', size)
else:
     print('the length of the chain is', length)
```
