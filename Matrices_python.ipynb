```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Altered code to find if det(A) = 2 
#code 2
n = 4
z = int((n**2)/2) + ((n**2) % 2 > 0) #the max number of 0s/1s in the matrix 
A = np.zeros([n,n],int)
A[:,:] =2 #ensures both u and v start at 0
uv = np.zeros(2, int)
#u = uv[0], v = uv[1]
def randmat(A,uv):
    uv = np.zeros(2, int)
    for i in range(n):
        for j in range(n):
            A[i,j] = random.randint(0, 1)
            uv[A[i,j]] +=1
            #my way of limiting the number of 0s and 1s
            #if u > z, the matrix will be 1s from that point
            if (uv[0] == z+1):
                A[i,j] = 1
                uv[0] = z
                uv[1] += 1
            #the same for if v > z    
            elif (uv[1] == z+1):
                A[i,j] = 0
                uv[0] +=1
                uv[1] = z
            j +=1    
            if j == n:
                i +=1
                j = 0
                if i == n:
                    return 

def detfinder(x):
    for i in range(1000):
        if x == np.linalg.det(A):
            return print(A)
        else:
            randmat(A,uv)
randmat(A,uv)
detfinder(0)
```

```python
#Altered code to keep u and v within n/2
#Also to plot det values
#code 3
n = 15
z = int((n**2)/2) + ((n**2) % 2 > 0) 
A = np.zeros([n,n],int)
A[:,:] =2
uv = np.zeros(2, int)
#u = uv[0], v = uv[1]
def randmat(A,uv):
    uv = np.zeros(2, int)
    for i in range(n):
        for j in range(n):
            A[i,j] = random.randint(0, 1)
            uv[A[i,j]] +=1
            if (uv[0] == z+1):
                A[i,j] = 1
                uv[0] = z
                uv[1] += 1
            #the same for if v reaches 5    
            elif (uv[1] == z+1):
                A[i,j] = 0
                uv[0] +=1
                uv[1] = z
            elif (abs(uv[1] - uv[0] > n/2)):
                uv[A[i,j]] = uv[A[i,j]] - 1
                A[i,j] = (A[i,j] + 1) % 2
                uv[A[i,j]] +=1    
            j +=1    
            if j == n:
                i +=1
                j = 0
                if i == n:
                    return 
randmat(A,uv)
Det = []
bigmatrix = np.zeros([n,n], int)
count = 0
def detcounter(Det, count, w):
    for i in range(500):
        randmat(A,uv)
        Det.append(round(np.linalg.det(A)))
        if (abs(np.linalg.det(A)) > w):
            count+=1
            for i in range(n):
                for j in range(n):
                    bigmatrix[i,j] = A[i,j]
            print(A, 'has a determinant of:', round(np.linalg.det(bigmatrix)))
    print('there are', count, 'matrices with det(A) >',detlimit,'  or det(A) < -',detlimit)
detlimit = 300
detcounter(Det, count,detlimit)
Min = min(Det)
Max = max(Det)
bins = np.arange(Min, Max+2)-0.5
fig, ax = plt.subplots(1,1)
ax.hist(Det, bins, density= True)
ax.set_xlim(Min-1,Max+1)
ax.set_xlabel('Det(A)')
ax.set_ylabel('probability mass function')
plt.show
```
