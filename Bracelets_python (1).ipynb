```python
#main code
#function to see the bracelet generated by any two numbers
n = 64
ans = np.zeros(n, int)
def bracelet(ans):
    for i in range(n):
        ans[i+2] = ans[i+1] + ans[i]
        ans[i+2] = ans[i+2] % 10
        print(ans[i])
        if (ans[i+1] == ans[0] and ans[i+2] == ans[1]):
            print('the length of the bracelet is:', i+1)
            break
def twonumb(x, y):
    ans[0] = x
    ans[1] = y
    return bracelet(ans)

twonumb(8,8)
```
